```python

import seaborn as sns
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
palette=['#ff3b30', '#007aff', '#4cd964','#ffcc00', '#5856d6', '#ff9500', '#5ac8fa']
markers=['o', '^', '<', 'd', '*','2']
sns.set_style("ticks")
import matplotlib as mpl
mpl.rcParams['lines.markersize'] = 6
mpl.rcParams['lines.linewidth']=1
rc("font", **{'family':'serif','serif':['Palatino'], 'size':14})
rc('text', usetex=True)
import os
```

```python

def plot_one_metric(total, name, title, metric_name, show_title=False, show_xlabel=True):
    GP=[abs(float(line.split("\t")[0])) for line in total.split("\n")]
    EGP=[abs(float(line.split("\t")[1])) for line in total.split("\n")]
    VPG=[abs(float(line.split("\t")[2])) for line in total.split("\n")]
    PQT=[abs(float(line.split("\t")[3])) for line in total.split("\n")]
    DSR=[abs(float(line.split("\t")[4])) for line in total.split("\n")]
    GPMELD=[abs(float(line.split("\t")[5])) for line in total.split("\n")]
    eureqa=[abs(float(line.split("\t")[6])) for line in total.split("\n")]
    df=pd.DataFrame({
        "model": EGP+GP+VPG+PQT+DSR+GPMELD+eureqa,
        "type": ['CVGP']*10+['GP']*10+['VPG']*10+['PQT']*10+['DSR']*10+['GPMeld']*10+['Eureqa']*10
    })
    

    plt.figure(figsize=(4.5,2.5))
    ax=sns.boxplot(data=df, y="model", x="type", 
                   showfliers=True, 
                   width=0.5,
                    palette=palette)
    if show_title==True:
        ax.set_title(title, fontsize=16)
    
    ax.set(ylabel=metric_name, xlabel="")
    plt.yscale('log', base=10)
    if show_xlabel==False:
        ax.set(xticklabels=[])
    plt.xticks(rotation=-24)
    fname = os.path.join(basepath, name+"_metric.pdf")
    plt.savefig(fname, bbox_inches='tight', pad_inches=0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df_sz = pd.DataFrame({'Ann1': (2 ** np.random.uniform(1, 20, 200)).astype(int),
                      'Ann2': (2 ** np.random.uniform(1, 20, 200)).astype(int),
                      'Ann3': (2 ** np.random.uniform(1, 20, 200)).astype(int),
                      'Ann4': (2 ** np.random.uniform(1, 20, 200)).astype(int)})

fig, ax = plt.subplots(figsize=(20, 4)) # very wide figure to accomodate 200 bars
sns.histplot(data=df_sz.melt(), x='value', hue='variable', bins=50,
             alpha=0.5, shrink=0.8, log_scale=True, multiple='dodge', ax=ax)
ax.legend_.set_title('') # remove the legend title (the name of the hue column)
ax.margins(x=0.01) # less spacing
plt.tight_layout()
plt.show()
```
